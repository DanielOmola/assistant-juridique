# 1 - INSTALLATION DES BIBLIOTHÈQUES

```python
# logger.info("🚀 Démarrage du backend")
# logger.info("📦 Installation des bibliothèques...")
# ============================================
# PRODUCTION-READY: Persistent venv with error handling
# ============================================
from google.colab import drive
drive.mount('/content/drive')

import os
import sys
import subprocess
from pathlib import Path

class ColabVenv:
    def __init__(self, drive_path='/content/drive/MyDrive/colab_venv'):
        self.venv_path = Path(drive_path)
        self.python_version = f"{sys.version_info.major}.{sys.version_info.minor}"
        self.site_packages = self.venv_path / f'lib/python{self.python_version}/site-packages'
        
    def setup(self):
        """Setup persistent venv if it doesn't exist"""
        if self.venv_path.exists():
            print("✅ Using existing persistent venv")
            return True
        
        print("📦 First-time setup: Creating persistent venv on Drive...")
        print("⏱️  This will take 2-3 minutes, but only once!")
        
        # Create venv
        subprocess.run([sys.executable, '-m', 'venv', str(self.venv_path)], check=True)
        
        # Get venv pip
        pip = self.venv_path / 'bin/pip'
        
        # Install all packages (including full langchain ecosystem)
        packages = [
            # Core
            'openai>=1.0.0',
            # Langchain ecosystem
            'langchain>=0.3.0',
            'langchain-community>=0.3.0',
            'langchain-core>=0.3.0',
            'langchain-openai>=0.2.0',
            'langchain-anthropic>=0.2.0',
            'langchain-google-genai>=0.1.0',
            # Document processing
            'pypdf>=3.0.0',
            'tiktoken>=0.5.0',
            'PyPDF2>=3.0.0',
            'python-docx>=1.0.0',
            # UI
            'ipywidgets>=7.7.0',
            # Other LLMs
            'anthropic>=0.18.0',
            'groq>=0.4.0',
            'mistralai>=0.1.0',
            'google-generativeai>=0.3.0',
        ]
        
        # Install everything at once (much faster)
        subprocess.run([str(pip), 'install'] + packages, check=True)
        
        print("✅ Persistent venv created successfully!")
        return True
    
    def activate(self):
        """Activate venv in current session"""
        if self.site_packages.exists():
            sys.path.insert(0, str(self.site_packages))
            print(f"✅ Venv activated: {self.site_packages}")
            return True
        else:
            print(f"⚠️  Site-packages not found at {self.site_packages}")
            return False
    
    def verify(self):
        """Verify critical imports"""
        try:
            from langchain_openai import ChatOpenAI
            from langchain_community.document_loaders import PyPDFLoader
            import openai, pypdf, tiktoken
            print("✅ All critical packages verified!")
            return True
        except ImportError as e:
            print(f"❌ Verification failed: {e}")
            return False

# Run setup
venv = ColabVenv()
venv.setup()
venv.activate()

# Install your local package
!pip install --no-deps -e /content/drive/MyDrive/assistant-juridique

# Verify everything works
venv.verify()
print("\n🎉 Your persistent environment is ready!")
print("✅ Bibliothèques installées")
```
